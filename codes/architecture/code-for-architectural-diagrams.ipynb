```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os

def generate_balanced_research_strategy_v34():
    save_dir = "results"
    if not os.path.exists(save_dir): os.makedirs(save_dir)
    save_path = os.path.join(save_dir, "SwinArchitecture.png")

    fig, ax = plt.subplots(figsize=(18, 28))
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.axis('off')

    # Professional business theme (light, muted, high readability)
    colors = {
        'input': '#F9FAFB',       # very light gray
        'backbone': '#F3F4F6',    # soft gray
        'phase1': '#EFF6FF',      # muted blue
        'phase2': '#F5F3FF',      # muted lavender
        'final': '#FFFBEB',       # warm off-white
        'border': '#9CA3AF',      # soft neutral gray
        'shadow': '#E5E7EB',      # lighter shadow
        'text': '#111827',        # dark gray (not harsh black)
        'subbox': '#FFFFFF'       # white
    }

    def draw_aligned_box(x, y, w, h, title, details=[], color='#FFFFFF', header_color=None):
        shadow = patches.Rectangle((x+0.2, y-0.2), w, h, facecolor=colors['shadow'], zorder=1)
        ax.add_patch(shadow)
        
        rect = patches.Rectangle((x, y), w, h, linewidth=1.2, edgecolor=colors['border'], 
                                 facecolor=color, zorder=2)
        ax.add_patch(rect)
        
        if header_color:
            header_rect = patches.Rectangle((x, y+h-4), w, 4, facecolor=header_color, 
                                            edgecolor=colors['border'], linewidth=1.2, zorder=3)
            ax.add_patch(header_rect)
            ax.text(x + w/2, y + h - 2, title, ha='center', va='center', 
                    fontweight='bold', fontsize=17, color=colors['text'], zorder=4)
        else:
            ax.text(x + w/2, y + h - 1.25, title, ha='center', va='center', 
                    fontweight='bold', fontsize=14, color=colors['text'], zorder=4)
        
        text_start_y = y + h - 3.2 if not header_color else y + h - 5.8
        for i, detail in enumerate(details):
            ax.text(x + 2.5, text_start_y - (i*2.2), f"• {detail}", ha='left', va='center', 
                    fontsize=13, color='#4B5563', zorder=4)

    draw_aligned_box(3, 85, 45, 12, "DATA INPUT STREAM", ["Balanced Augmentation Pipeline", "Effective Batch: 64 (16 x 4)", "Resolution: 224x224 RGB"], colors['input'], '#BFDBFE')
    draw_aligned_box(52, 85, 45, 12, "PRETRAINED ENGINE", ["Swin-B (Patch 4 / Win 7)", "ImageNet-22k Feature Space", "1024-Dimensional Latents"], colors['backbone'], '#FDE68A')

    draw_aligned_box(8, 57, 84, 24, "PHASE 1: WARMUP & STABILIZATION (Epochs 1-4)", 
                    ["Gradient Restriction: Stage < 2", "Linear LR Ramp: 1e-6 → 8e-5", 
                     "Freeze stages to stabilize MLP Head", "Optimizer: AdamW (Weight Decay: 0.02)"], colors['phase1'], '#BFDBFE')
    
    draw_aligned_box(13, 59.5, 22, 8, "Freezing", ["Locked: Stage 0-1", "Active: Stage 2-3"], colors['subbox'])
    draw_aligned_box(39, 59.5, 22, 8, "MLP Head", ["GELU / LayerNorm", "Dropout: 0.25"], colors['subbox'])
    draw_aligned_box(64, 59.5, 22, 8, "LR Policy", ["4 Epoch Warmup", "Peak: 8.0e-5"], colors['subbox'])

    draw_aligned_box(8, 19, 84, 34, "PHASE 2: GLOBAL FINE-TUNING (Epochs 5-30)", 
                    ["Global Parameter Unfreezing", "Cosine Annealing Decay Policy", 
                     "Backbone Differential LR: 0.08x", "HybridMix (Mixup/CutMix) enabled", 
                     "Label Smoothing: 0.05", "Early Stopping Patience: 7", 
                     "Stochastic Depth Rate: 0.15", "Gradient Clipping Norm: 1.0"], colors['phase2'], '#C7D2FE')
    
    draw_aligned_box(18, 21.5, 30, 8, "Full Optimization", ["End-to-End Tuning", "Joint Weight Updates"], colors['subbox'])
    draw_aligned_box(54, 21.5, 30, 8, "LR Schedule", ["Cosine Decay Active", "Min LR: 1.0e-6"], colors['subbox'])

    draw_aligned_box(25, 3, 50, 12, "TCMP-300 BENCHMARK TARGETS", ["Target Accuracy: > 92%", "Generalization Gap: < 2%", "Optimized FP16 Inference"], colors['final'], '#FDE68A')

    arr = dict(arrowstyle='-|>', color=colors['border'], lw=1.8, mutation_scale=18)
    ax.annotate("", xy=(52, 91), xytext=(48, 91), arrowprops=arr)
    ax.annotate("", xy=(50, 81), xytext=(50, 85), arrowprops=arr)
    ax.annotate("", xy=(50, 53), xytext=(50, 57), arrowprops=arr)
    ax.annotate("", xy=(50, 15), xytext=(50, 19), arrowprops=arr)

    plt.savefig(save_path, bbox_inches='tight', dpi=300)
    plt.show()

generate_balanced_research_strategy_v34()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os

def generate_tcmp300_research_final_clean():
    save_dir = "results"
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    save_path = os.path.join(save_dir, "SwinMLEnsembleArchitecture.png")

    fig, ax = plt.subplots(figsize=(26, 14))
    ax.set_xlim(0, 155)
    ax.set_ylim(0, 95)
    ax.axis('off')

    colors = {
        'input': '#F9FAFB',
        'backbone': '#EFF6FF',
        'hooks': '#E0E7FF',
        'fusion': '#ECFDF5',
        'learners': '#F5F3FF',
        'meta': '#FEF3C7',
        'final': '#FFFFFF',
        'border': '#6B7280',
        'shadow': '#D1D5DB',
        'text': '#111827'
    }

    def draw_component(x, y, w, h, title, subtitle="", color=colors['input'], text_size=11, title_bold=False, align_left=False):
        offset = 0.3
        shadow_rect = patches.FancyBboxPatch((x + offset, y - offset), w, h, 
                                            boxstyle="round,pad=0.3", 
                                            linewidth=0, facecolor=colors['shadow'], zorder=1)
        ax.add_patch(shadow_rect)
        
        rect = patches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.3", 
                                      linewidth=1.2, edgecolor=colors['border'], 
                                      facecolor=color, zorder=2)
        ax.add_patch(rect)
        
        weight = 'bold' if title_bold else 'normal'
        ax.text(x + w/2, y + h*0.75 if subtitle else y + h/2, title, ha='center', va='center', 
                fontweight=weight, fontsize=text_size, color=colors['text'], zorder=3, wrap=True)
        
        if subtitle:
            if align_left:
                ax.text(x + w*0.1, y + h*0.35, subtitle, ha='left', va='center', 
                        fontweight='normal', fontsize=text_size-2.5, color='#4B5563', zorder=3, wrap=True)
            else:
                ax.text(x + w/2, y + h*0.35, subtitle, ha='center', va='center', 
                        fontweight='normal', fontsize=text_size-2, color='#4B5563', zorder=3, wrap=True)

    draw_component(2, 40, 12, 16, "TCMP-300 CHINESE\nHERBAL DATASET", "(300 Herb Classes)")
    draw_component(16, 40, 11, 16, "TTA PIPELINE", "10-View Transform\n(5-Crop + HFlip)")
    draw_component(30, 28, 18, 40, "SWIN TRANSFORMER\nBACKBONE", "Hierarchical Architecture\n(Patch Embed + Stage 1-4)", colors['backbone'])
    draw_component(31, 75, 16, 10, "LAYER 1 HOOK", "Spatial Detail Features", colors['hooks'])
    draw_component(31, 12, 16, 10, "LAYER 3 HOOK", "Global Semantic Feat.", colors['hooks'])
    draw_component(52, 40, 14, 16, "LATENT FUSION", "MS-Concat + Pooling\nPCA (768 Dimensions)", colors['fusion'])

    y_coords = [82, 72, 62, 52, 42, 32, 22]
    clfs = [("SVM-RBF", "C=25.0"), ("LIGHTGBM", "800 Trees"), ("XGBOOST", "200 Trees"), 
            ("MLP HEAD", "Deep MLP"), ("EXTRATREES", "600 Est."), ("LOGISTIC REG", "L2 Reg"), ("KNN", "Cosine")]
    for i, (name, config) in enumerate(clfs): 
        draw_component(72, y_coords[i], 13, 7, name, config, colors['learners'], text_size=9)

    draw_component(92, 60, 11, 18, "STACKING", "Meta-Learner\n(LogReg on OOF)", colors['meta'])
    draw_component(92, 22, 11, 18, "WEIGHTED", "Soft-Vote\n(Weight ∝ Acc²)", colors['meta'])
    
    draw_component(112, 40, 13, 16, "ALPHA BLEND", r"$\alpha$-Sweep Blend" + "\n(Swin + ML Ensemble)", colors['final'])
    
    herbal_classes = "1. Veronica persica\n2. Lysimachia clethr.\n3. Alangium chinense\n...\n300. Callicarpa bodin."
    draw_component(131, 40, 13, 16, "OUTPUT CLASSES", herbal_classes, colors['final'], text_size=11, align_left=True)

    std_arrow = dict(arrowstyle='-|>', color=colors['border'], lw=1.3, mutation_scale=15)
    
    ax.annotate("", xy=(16, 48), xytext=(14, 48), arrowprops=std_arrow)
    ax.annotate("", xy=(30, 48), xytext=(27, 48), arrowprops=std_arrow)
    ax.annotate("", xy=(52, 48), xytext=(48, 48), arrowprops=std_arrow)
    
    for y in y_coords: 
        ax.annotate("", xy=(72, y + 3.5), xytext=(66, 48), arrowprops=std_arrow)
        
    for y in y_coords:
        ax.annotate("", xy=(92, 69), xytext=(85, y + 3.5), arrowprops=dict(arrowstyle='-|>', color=colors['border'], lw=1.0, alpha=0.5, mutation_scale=15))
        ax.annotate("", xy=(92, 31), xytext=(85, y + 3.5), arrowprops=dict(arrowstyle='-|>', color=colors['border'], lw=1.0, alpha=0.5, mutation_scale=15))
    
    ax.annotate("", xy=(112, 49), xytext=(103, 69), arrowprops=std_arrow)
    ax.annotate("", xy=(112, 47), xytext=(103, 31), arrowprops=std_arrow)
    
    ax.plot([39, 39], [68, 75], color=colors['border'], lw=1.0, ls='--', zorder=0)
    ax.plot([39, 39], [22, 28], color=colors['border'], lw=1.0, ls='--', zorder=0)
    
    ax.annotate("", xy=(131, 48), xytext=(125, 48), arrowprops=std_arrow)

    plt.savefig(save_path, bbox_inches='tight', dpi=300)
    plt.show()

generate_tcmp300_research_final_clean()
```
